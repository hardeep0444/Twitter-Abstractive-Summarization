```python
df=pd.read_csv("/kaggle/input/chat-gpt-dataset/dataset.csv")
```

```python
df
```

```python
df = df.rename(columns={"x":"source_text", "y":"target_text"})
df = df[['source_text', 'target_text']]
```

```python
df
```

```python
df['source_text'] = "summarize: " + df['source_text']
```

```python
df
```

```python
train_df, test_df = train_test_split(df, test_size = 0.15)
train_df.shape, test_df.shape
```

```python
from simplet5 import SimpleT5
model = SimpleT5()
model.from_pretrained(model_type="t5", model_name="t5-base")
```

```python
from tqdm import tqdm
import torch
!pip install rouge_score
```

```python
model.device= torch.device("cpu")
```

```python
pred=[]
for i in tqdm(test_df['source_text'],total=len(test_df)):
    pred.append(model.predict(i))
```

```python
ref_text=[]
for i in test_df['target_text']:
    ref_text.append(i)
```

```python
from datasets import load_dataset, load_metric
rouge_metric = load_metric('rouge')
rouge_metric.add_batch(predictions=pred, references=ref_text)
        
#  Finally compute and return the ROUGE scores.
score = rouge_metric.compute()
rouge_names = ["rouge1", "rouge2", "rougeL", "rougeLsum"]
rouge_dict = dict((rn, score[rn].mid.fmeasure ) for rn in rouge_names )

pd.DataFrame(rouge_dict, index = ['pegasus'])
```

```python
model.train(train_df=train_df,
            eval_df=test_df, 
            source_max_token_len=1024, 
            target_max_token_len=150, 
            batch_size=4, max_epochs=5, use_gpu=True)
```
